# Constraint Extractor: Rule-Based Time and Effort Detection

This notebook demonstrates the `ConstraintExtractor` module, which uses keyword lists and regex patterns to extract time and effort constraints from free-text recipe prompts.

```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))

from pantry.constraint_extractor import ConstraintExtractor

extractor = ConstraintExtractor()
```

## Keyword Lists and Regex Patterns

Below are the rules the extractor uses.

```python
print("=== Time Extraction Rules ===\n")
print(f"Minutes regex: {extractor._MINUTES_RE.pattern}")
print(f"Hours regex:   {extractor._HOURS_RE.pattern}")
print(f"\nPhrase patterns:")
for pattern, value in extractor._TIME_PHRASES:
    print(f"  {pattern.pattern!r} -> {value}")
print(f"\nTime keywords:")
for kw, val in extractor._TIME_KEYWORDS.items():
    print(f"  {kw!r:15s} -> {val}")

print("\n=== Effort Extraction Rules ===\n")
print("Effort keywords:")
for kw, val in extractor._EFFORT_KEYWORDS.items():
    print(f"  {kw!r:15s} -> {val}")
```

## Extraction Results on Example Prompts

```python
examples = [
    "something quick, 20 minutes",
    "30 minutes max",
    "easy weeknight dinner",
    "chicken and rice",
    "quick and easy pasta",
    "under an hour",
    "2 hours slow roast",
    "I want a simple beginner-friendly meal",
    "give me a gourmet elaborate dinner",
    "fast 15 min stir fry",
    "moderate effort, about 45 minutes",
    "lazy Sunday brunch, nothing complex",
    "advanced French cuisine, all day project",
]

print(f"{'Prompt':<50} {'Time':<10} {'Effort':<10}")
print("-" * 70)
for prompt in examples:
    result = extractor.extract(prompt)
    t = result["time"] or "-"
    e = result["effort"] or "-"
    print(f"{prompt:<50} {t:<10} {e:<10}")
```

## Edge Cases

```python
edge_cases = [
    ("", "Empty string"),
    ("90 minutes", "Minutes over 60 -> long"),
    ("under 30 minutes please", "Under 30 phrase -> short"),
    ("QUICK AND EASY", "All uppercase"),
    ("speedy effortless snack", "Multiple keywords (first match wins)"),
    ("I have 5 mins", "Very short time -> short"),
    ("intermediate difficulty", "Medium effort keyword"),
]

print(f"{'Input':<40} {'Note':<35} {'Time':<8} {'Effort':<8}")
print("-" * 91)
for text, note in edge_cases:
    result = extractor.extract(text)
    t = result["time"] or "-"
    e = result["effort"] or "-"
    display = text if text else "(empty)"
    print(f"{display:<40} {note:<35} {t:<8} {e:<8}")
```
